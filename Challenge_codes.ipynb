```python
### Importing pandas as pd, regex as re.

import pandas as pd
import re
import math

### Creating an header for our table

header_we = ['Account_Number', 'Account_Name', 'Transaction_Fee', 'Phone_Number']

### Importing csv as df

df = pd.read_csv('accounts.csv',  delimiter = "|", names = header_we)
```

```python
### printing csv denote df

df
```

```python
### Using lambda format the Account Name by adding leading zeros at and making sure its 10 digit long.

df['Account_Number']=df['Account_Number'].apply(lambda x: '{0:0>10}'.format(x))
```

```python
### Result 

df['Account_Number']
```

```python
### Using regex and stip and replace method to delete unecessary charaters in df['Acoount_Number'] as s except (, - space)

df['Account_Name'] = df['Account_Name'].str.replace("(\(.*?\))", "").str.strip("'''")
```

```python
df['Account_Number']
```

```python
### Round transaction fee to 2 decimal place and making sure its int&float.

df['Transaction_Fee'] = df['Transaction_Fee'].round(decimals=2)

### Result

df['Transaction_Fee']
```

```python
### Eliminating characters in phone_Number with regex and converting it to a string formatted as ###-###-####
  
df['Phone_Number'] = df['Phone_Number'].map(lambda x: re.sub(r'\W+', '', x))

df['Phone_Number']
```

```python
#### converting the number to ###-###-#### format

contactphone=list()
for phone_no in df['Phone_Number']:
    contactphone.append("%c%c%c-%c%c%c-%c%c%c%c" % tuple(map(ord,list(str(phone_no)[:10]))))
contactphone = pd.DataFrame(contactphone)

### Assigning contact phone to [Phone_Number]

contactphone
```

```python
df['Phone_Number'] = contactphone
df['Phone_Number']
```
